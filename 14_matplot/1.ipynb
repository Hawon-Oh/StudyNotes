### matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### 기본 사용법

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y)
plt.show()
```

##### Figure | Axes

- Figure: 전체 그림이 그려지는 컨테이너
- Axes: 그래프가 실제로 그려지는 영역(좌표, 데이터)

```python
fig = plt.figure(figsize=(6, 4), facecolor='green')
ax = plt.gca()
ax.plot([1,2,3],[4,5,6])
plt.title('hello matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,6))
print(fig)
print(ax)
plt.show()
```
