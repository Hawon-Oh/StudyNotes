##### EDA 전에 전처리
##### EDA 이후 데이터 파악(데이터 분포, 비즈니스 특성 등)을 하고 또 전처리

##### EDA (Exploratory Data Analysis, 탐색적 요인분석)

### EDA 절차

1. 데이터 로드
- 다양한 소스에서 데이터를 로드한다. (파일, 데이터베이스, API 등에서 데이터를 불러온다.)
- 데이터가 로드된 후에는 구조를 확인하기 위해 첫 몇 줄의 데이터를 출력해 본다.
2. 데이터 구조 및 기초 통계 확인
- 데이터의 구조를 파악하기 위해 컬럼 정보와 데이터 타입을 확인한다.
- 데이터의 기본 통계 정보를 출력하여 각 변수의 분포와 특성을 살펴본다.
- df.head(), df.info(), df.describe() 등의 함수를 사용한다.
3. 결측치 및 이상치 탐색
- 결측치(NaN) 값이 존재하는지 확인하고, 이를 처리하는 방법을 결정한다.
- 데이터에 존재하는 이상치(Outlier)를 탐지하고, 이를 어떻게 처리할지 결정한다.
- df.isnull().sum(), df.boxplot() 등의 함수를 활용한다.
4. 데이터 시각화를 통한 탐색
- 데이터를 시각화하여 변수 간의 관계, 분포 등을 파악한다.
- 히스토그램, 박스플롯, 상관관계 행렬 등 다양한 그래프를 통해 데이터의 특성을 시각적으로 확인한다.
- sns.countplot(), sns.heatmap() 등의 함수를 사용한다.
5. 데이터 정제 및 전처리
- 필요 없는 변수나 중복 데이터를 제거한다.
- 범주형 데이터를 처리하거나, 스케일링 및 정규화를 통해 모델에 적합한 형태로 데이터를 변환한다.
- df.drop(), df.fillna(), pd.get_dummies() 등의 함수를 활용한다.
6. 데이터 변환 및 피처 엔지니어링
- 새로운 피처를 생성하거나 기존 피처를 변환하여 분석에 적합한 형태로 데이터를 조정한다.
- 로그 변환, 다항식 피처 추가 등 다양한 기법을 통해 데이터를 변환할 수 있다.
- np.log(), PolynomialFeatures() 등의 함수를 활용한다.
7. 데이터 분할
- 학습용과 테스트용 데이터로 분할한다.
- 이 과정은 모델을 평가하고 성능을 검증하는 데 중요한 단계이다.
- train_test_split() 함수를 사용한다.

**컬럼 설명** 
- PassengerId - 승객 고유번호
- Survived - 생존유무(target 값) (0 = 사망, 1 = 생존)
- Pclass - 티켓 클래스 (1 = 1st, 2 = 2nd, 3 = 3rd)
- Name - 탑승객 성명
- Sex - 성별 (male: 남성, female: 여성)
- Age - 나이(세)
- SibSp - 함께 탑승한 형제자매, 배우자 수 총합
- Parch - 함께 탑승한 부모, 자녀 수 총합
- Ticket - 티켓 넘버
- Fare - 탑승 요금
- Cabin - 객실 넘버
- Embarked - 탑승 항구 (C: Cherbourg, Q: Queenstown, S: Southampton)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 데이터 로드
# 2. 데이터 구조 및 기초 통계 확인
# 3. 결측치 및 이상치 탐색
# 4. 데이터 시각화를 통한 탐색
# 5. 데이터 정제 및 전처리
# 6. 데이터 변환 및 피처 엔지니어링
# 7. 데이터 분할
```

```python
# 1. 데이터 로드
df = pd.DataFrame(pd.read_csv('../../01_python_basic/data/titanic.csv'))
df.head()
```

```python
# 2. 데이터 구조 및 기초 통계 확인
df.info()
```

```python
df.describe()
```

```python
# 3. 결측치
df.isnull().sum()
# df.count()
```

```python
# 이상치 탐색
df['Age'].plot(kind='box')
df[df['Age'] < 1]
```

```python
# 데이터 시각화를 통한 탐색
sns.countplot(data=df, x='Survived', hue='Sex')
plt.show()
```

```python
df['Age'].hist(bins=10)
```

```python
# numeric_only은 숫자형 데이터만 쓴다는 뜻
# 현재 성별은 텍스트라서 히트맵에 표시안됨
corr_matrix = df.corr(numeric_only=True)

# annot 히트맵에 숫자도 표시함
sns.heatmap(corr_matrix, annot=True)
```

```python
# 5. 데이터 정체 및 전처리
#
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'].isnull().sum()
```

```python
# 크게 상관없을 데이터는 drop
df = df.drop('Cabin', axis=1)
```

```python
# 비슷한 의미를 지닌 데이터끼리 합침
df['Family'] = df['SibSp'] + df['Parch']
df = df.drop(['SibSp', 'Parch'], axis=1)
df
```

##### 한쪽으로 치우친 데이터는 학습에 안 좋음.
    - 분포를 최대한 정규화시켜야 함

```python
df['Fare'].hist(bins=20)
plt.show()

df['LogFare'] = df['Fare'].apply(lambda x: np.log1p(x))
df['LogFare'].hist()
```

```python
# 6. 데이터 변환 및 피처 엔지니어링
from sklearn.preprocessing import LabelEncoder

# str 성별 -> numeric 성별
encoder = LabelEncoder()
# fit은 계산하고  transform은 변환한다는 뜻
df['Sex'] = encoder.fit_transform(df['Sex'])
df.head(1)
```

```python
# 7. 데이터 분할
from sklearn.model_selection import train_test_split

X_data = df.drop('Survived', axis=1)
Y_data = df['Survived']
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data)
```
