# GMM (Gaussian Mixture Model)

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
X, _ = make_classification(
    n_samples=300,          # 데이터 개수
    n_features=2,           # 특성 개수
    n_informative=2,        # 유의미한 특성 개수
    n_redundant=0,          # 중복특성 개수
    n_clusters_per_class=1, # 클래스별 클러스터
    n_classes=2,            # 클래스(레이블) 수
    random_state=42
)

plt.scatter(X[:, 0], X[:, 1], s=50, c='gray', marker='o', edgecolors='k')
plt.title('scikit-learn making data')
plt.show()
```

```python
# GMM 적용 및 군집화 결과 시각화
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X)

labels = gmm.predict(X)

plt.scatter(X[:, 0], X[:, 1], s=50, c=labels, marker='o', edgecolors='k')
plt.title('GMM cluster result')
plt.show()
```

```python
# 군집 포함 확률 시각화 (확률이 높다고 예측한대로 시각화)
probs = gmm.predict_proba(X)

plt.scatter(X[:, 0], X[:, 1], s=50, c=probs.max(axis=1), marker='o', edgecolors='k')
plt.title('GMM cluster rate')
plt.colorbar(label='rate')
plt.show()
# probs
```
