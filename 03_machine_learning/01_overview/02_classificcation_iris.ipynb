## [실습] 붓꽃 다중분류

붓꽃은 세포핵을 둘러싼 꽃받침(sepal)과 꽃잎(petal)으로 구성되어 있으며, 이러한 특징을 기반으로 붓꽃의 품종을 식별할 수 있습니다.

- Setosa (세토사): Setosa는 붓꽃 중에서 가장 작은 꽃잎과 꽃받침을 가지고 있습니다.
꽃잎과 꽃받침이 비교적 짧고 뾰족한 모습을 갖고 있으며, 주로 흰색 또는 연한 분홍색을 띠고 있습니다.

- Versicolor (버시컬러): Versicolor는 Setosa보다 크고 긴 꽃잎과 꽃받침을 가지고 있습니다.
꽃잎의 색은 보통 연한 보라색이며, 중간 크기의 붓꽃입니다.

- Virginica (버지니카): Virginica는 붓꽃 중에서 가장 크고 긴 꽃잎과 꽃받침을 가지고 있습니다.
꽃잎의 색은 주로 짙은 보라색이며, 다른 품종들에 비해 상대적으로 더 큰 크기를 갖고 있습니다.

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
```

```python
print(iris_data.keys())
print(iris_data.target_names)
print(iris_data.target)
print(iris_data.feature_names)
print(iris_data.data)
```

##### 최근접이웃 분류 모델로 훈련/평가 해보기

```python
iris_data.data
```

```python
import numpy as np
import pandas as pd
iris_df = pd.DataFrame(iris_data.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
iris_df['label'] = iris_data.target
iris_df.describe()
```

```python
iris_df[iris_df['petal_length'] < 1.6].label
```

```python
# 데이터 전처리
# 1. 입력-라벨 데이터 분리
input_data, label_data = iris_df.drop(columns=['label']), iris_df.label
print(input_data.shape, label_data.shape)

# 2. 훈련-테스트 데이터 분리
from sklearn.model_selection import train_test_split

train_input, test_input, train_label, test_label = train_test_split(input_data, label_data, test_size=0.25, random_state=42)


# 3. 정규화
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
standard_scaler.fit(train_input) # 훈련 데이터 기준 평균/표준편차 작업
train_scaled = standard_scaler.transform(train_input) # 훈련 데이터 정규화 반영
test_scaled = standard_scaler.transform(test_input)   # 테스트 데이터 정규화 반영
```

```python
from sklearn.neighbors import KNeighborsClassifier

# 훈련
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_label)

# 평가
score = kn.score(test_scaled, test_label)
print('score: ', score)

# 평가지표
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(kn.predict(test_scaled), test_label)
print('accuracy:', accuracy)
```
