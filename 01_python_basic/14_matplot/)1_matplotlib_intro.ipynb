##### Figure | Axes

- Figure: 전체 그림이 그려지는 컨테이너
- Axes: 그래프가 실제로 그려지는 영역(좌표, 데이터)

```python
fig = plt.figure(figsize=(6, 4), facecolor='green')
ax = plt.gca()
ax.plot([1,2,3],[4,5,6])

plt.title('hello matplotlib')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(10,6))
print(fig)
print(ax)
plt.show()
```

```python
# List
x = range(10)
y = range(10)
plt.figure(figsize=(3, 2))
plt.plot(x, y)
plt.show()
```

```python
# 4. DataFrame
x = range(10)
y = range(10)
x = pd.DataFrame({'X': x})
x = pd.DataFrame({'Y': y})
plt.figure(figsize=(2,3))
plt.plot(x, y)
plt.show()
```

##### plot 메서드의 매개변수

```python
plt.figure(figsize=(3, 2))
plt.plot(
    x,
    y,
    color='blue',
    marker='o',
    markersize=5,
    linestyle='--',
    linewidth=2

)
```

```python
# label: 한글폰트 지원 X -> font manager로 설정해야 함
plt.plot(x, y)
plt.title('제목')
plt.xlabel('시간')
plt.ylabel('점수')
plt.show()
```

```python
plt.Figure(figsize=(3, 4))

x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.xticks(ticks=np.arange(0, 101, 5), rotation=90)
plt.yticks(rotation=45)
plt.yticks()
plt.plot(x, y)
plt.show()
```

```python
# lim (plot show되는 범위 지정)
plt.Figure(figsize=(3, 2))
x = np.arange(1, 101)
y = np.linspace(2, 200, 100)

plt.xlim(1, 200)
plt.ylim(2, 100)

plt.plot(x, y)
plt.show()
```

```python
# legend(범례)
plt.Figure(figsize=(3, 2))
x = np.arange(1, 101)
y = np.linspace(2, 200, 100)
y2 = np.linspace(3,300, 100)

plt.plot(x,y,label='시간')
plt.plot(x,y2, label='공간')
plt.legend() # default는 best(최적위치 찾음)
# plt.legend(loc='lower right')
# plt.legend(loc='center')
plt.show()
```

##### 그래프를 여러 개 그리기

```python
plt.Figure(figsize=(3, 2))
x = np.arange(1, 11)
y = x * 3
y2 = x * 2
plt.plot(x, y, label='선', color='red', marker='o')
plt.bar(x, y2, label='바')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# subplots 이용 -> 여러 그래프 그리기
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

x1 = np.arange(1, 11)
y1 = x1 * 2
x2 = np.arange(11, 20)
y2 = x2 * 2

ax[0].plot(x1, y1, color='red', label='선')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('Y1')
ax[0].set_title('빨간 선')
ax[0].legend()
ax[0].grid()

ax[1].plot(x2, y2, color='blue', label='선')
ax[1].set_xlabel('X2')
ax[1].set_ylabel('Y2')
ax[1].set_title('연두 막대')
ax[1].legend()
ax[1].grid()



plt.show()
```
