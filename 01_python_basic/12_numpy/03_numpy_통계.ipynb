### 통계 관련 numpy

```python
import numpy as np
```

##### 중심 경향 측도

```python
height = np.array([160, 170, 180, 150, 155, 165, 175, 185, 180, 175, 170, 173, 164, 189, 175])
```

```python
# 평균값
print(np.mean(height))
```

```python
# 중위값
print(np.sort(height))
print(np.median(height))
```

```python
# 최빈값
print(np.unique(height, return_counts=True))
values, count = np.unique(height, return_counts=True)
```

```python
max_count = np.max(count)
print(count == max_count)
print(values[count == max_count])
```

## 산포 측도

```python
# 분산 (variance)
#  Σ(값 - 평균)^ / N
print(np.mean(np.square(height - np.mean(height)))) # 또는
var1 = np.var(height)
print(var1)
```

```python
# 표준 편차 (standard deviation)
# 분산의 제곱근
std1 = np.sqrt(var1)
print(std1)
```

```python
import matplotlib.pyplot as plt
```

```python
# 정규분포
# np.random.randn(shape)
# 평균이 0, 표준편차가 1인 난수 출력 함수
np.random.randn()
```

```python
arr = np.random.randn(1000)
plt.hist(arr, bins=100)
plt.show()
```

```python
# np.random.normal(loc, scale, size)
# - loc: 평균
# - scale: 표준편차
# - size: 갯수 or shape
arr = np.random.normal(10, 1, 1000)
print(arr[:50])

plt.hist(arr, bins=100)
plt.show()
```

```python
# 균등분포
# np.random.rand(num)
# 균등하게 0 ~ 1의 난수 생성
# - num: 개수 또는 shape
arr = np.random.rand(1000)

plt.hist(arr, bins= 50)
plt.show()
```

```python
# 균등분포
# np.random.randint(start, end, size)

np.random.seed(0) # 난수생성 씨드 고정
arr = np.random.randint(1, 100, (100, 20))

print(arr[:10])
plt.hist(arr, bins=5)
plt.show()
```

```python
arr = np.array([0, 1, 3, 2, 4, 6, 5, 7, 9, 8])
print(np.sort(arr, axis=None))
print(np.sort(arr, axis=-1))
print(arr.ndim)
```

```python
# np.argsort()
# - 배열을 정렬한다면 원래 인덱스는 어떻게 될지를 반환

np.argsort(arr)
```

```python
names = np.array(['리사', '로제', '파스타'])
album_sales = np.array([900, 870, 2000])

names[np.argsort(album_sales)[::-1]]
```
