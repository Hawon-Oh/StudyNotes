```python
# tsv파일은 구분자가 쉼표가 아니라 탭임.
file_path = "./data/users.tsv"
df = pd.read_csv(file_path, sep='\t')
df
```

```python
# column명은 없지만 header=None으로 숫자 출력 (아니면 첫번째 row가 col명이 됨)
# names를 쓰면 col명을 바꿀 수 있음
file_path = "./data/headless_users.csv"
df = pd.read_csv(file_path,
                 header=None, # 첫 줄에 있는 데이터가 header가 되지 않게 함
                 names=['id', 'first_name', 'last_name', 'email', 'gender', 'ip_address'])
df
```

```python
df.info()
```

```python
df.to_csv('./data/users2.csv') # 기본값: index=True, header=True
df.to_csv('./data/users2.csv', index=False, header=False)
```
