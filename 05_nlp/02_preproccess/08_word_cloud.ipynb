### 텍스트 전처리


```python
import os, sys, glob, warnings, random, json
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
```

```python
spam_df = pd.read_csv('./sms_spam.csv')
spam_df.head()
```

```python
corpus = spam_df['CN'][0]
```

```python
okt = Okt()
nouns = okt.nouns(corpus)
nouns
```

```python
word_count = Counter(nouns)
word_count
```

```python
ko_stopwords = ['및', '더', '수']
word_count = {word: count for word, count in word_count.items() if word not in ko_stopwords}
word_count
```

```python
wordcloud = WordCloud(
    width=800,
    height=800,
    font_path='C:\\Windows\\SysWOW64\\Fonts\\NanumGothic.ttf',
    background_color='white'
).generate_from_frequencies(word_count)

plt.figure(figsize=(10, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
